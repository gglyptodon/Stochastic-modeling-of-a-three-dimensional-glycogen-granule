```python
def print_overview(g):
    t = "substrate for"
    s = f"{g}\n"
    s += f"{t} elongation: {g.find_chains_for_gs()}\n"
    s += f"{t} branching: {g.find_chains_for_gbe()}\n"
    s += f"{t} reduction: {g.find_chains_for_gp()}\n"
    s += f"{t} debranching: {g.find_chains_for_gde()}\n"
    print(s)
```
